```python
import pandas as pd
import matplotlib.pyplot as plt

# Function to read CSV and plot the GPU power usage
def plot_gpu_power(file_name):
    # Read the CSV file using pandas
    data = pd.read_csv(file_name)

    # Plotting the GPU power usage
    plt.figure(figsize=(10, 6))
    plt.plot(data['Time'], data['GPU Power (W)'], label=f'GPU Power ({file_name})')
    plt.title(f'GPU Power Usage Over Time for {file_name}')
    plt.xlabel('Time (seconds)')
    plt.ylabel('GPU Power (W)')
    plt.grid(True)
    plt.legend()
    plt.show()

# Function to calculate total energy consumption
def calculate_total_energy(file_name):
    # Read the CSV file
    data = pd.read_csv(file_name)

    # Assuming the time interval is constant and the same as the first interval
    if len(data['Time']) > 1:
        time_interval = data['Time'][1] - data['Time'][0]
    else:
        print("Not enough data points to calculate energy consumption.")
        return 0

    # Calculate total energy consumption in joules
    total_energy = sum(data['GPU Power (W)'] * time_interval)  # Power(W) * Time(s) = Energy(J)
    return total_energy

# Main function to run the applications, collect GPU power data, and plot results
def main():
    # Names of the CSV files
    files = ['./gpu_power_results_v1.csv', './gpu_power_results_v2.csv']

    # Plot power usage and calculate energy
    for file_name in files:
        plot_gpu_power(file_name)
        total_energy = calculate_total_energy(file_name)
        print(f"Total GPU energy consumption for {file_name}: {total_energy:.2f} Joules")

if __name__ == "__main__":
    main()
```
